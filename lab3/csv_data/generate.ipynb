```python
import numpy as np
import random
import pandas as pd
from math import exp

def gen_ab(a,b):
    x=np.linspace(0,50,num=100)
    y=np.zeros(100)
    for i in range(100):
        factor=random.gauss(1,0.2)
        y[i]=(x[i]*a+b)*factor
    return x,y

def gen_ab_exp(a,b):
    x=np.linspace(0,50,num=100)
    y=np.zeros(100)
    for i in range(100):
        factor=random.gauss(1,0.2)
        y[i]=(exp((a*x[i]+b)*factor/10))
    return x,y
```

```python
a=3
b=6
for (a,b) in [
    (3,6),(1,8),(4,9)]:
    time,y=gen_ab(a,b)
    df=pd.DataFrame({
        "time":time,
        "value":y})
    filename=f"linear_a_{a}_b_{b}.csv"
    df.to_csv(filename,
            sep=";",float_format="%.3f",index=False)

    df.plot(x="time")
```

```python
a=3
b=6
for (a,b) in [
    (3,6),(1,8),(4,9)]:
    time,y=gen_ab_exp(a,b)
    df=pd.DataFrame({
        "time":time,
        "value":y})
    filename=f"exp_a_{a}_b_{b}.csv"
    df.to_csv(filename,
            sep=";",float_format="%.3f",index=False)

    df.plot(x="time",logy=True)
    df.plot(x="time")
```
